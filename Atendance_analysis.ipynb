```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Load Dataset
df = pd.read_excel("attendance.xlsx")

# Total class duration
TOTAL_CLASS_TIME = 120

# Calculate Percentage
df["PERCENTAGE"] = (
    df["DURATION(MIN)"] /
    TOTAL_CLASS_TIME
) * 100

# Calculate Status
df["STATUS"] = df["DURATION(MIN)"].apply(
    lambda x: "Present" if x >= 60 else "Absent"
)

# Calculate Result
df["RESULT"] = df["PERCENTAGE"].apply(
    lambda x: "Qualified"
    if x >= 80
    else "Not Qualified"
)

# Label Encoding
df["LABEL"] = df["RESULT"].map(
    {
        "Qualified":1,
        "Not Qualified":0
    }
)

# Features and Target
X = df[["DURATION(MIN)", "PERCENTAGE"]]
y = df["LABEL"]

# Train Model
model = DecisionTreeClassifier()

model.fit(X, y)

# Prediction
predictions = model.predict(X)

# Convert back to text
df["ML_RESULT"] = [
    "Qualified"
    if p == 1
    else "Not Qualified"
    for p in predictions
]

# Display Data
display(df)

# Save Output
df.to_excel(
    "attendance_prediction.xlsx",
    index=False
)

print("File Saved Successfully")
```
